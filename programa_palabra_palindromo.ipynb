**Determinar si una palabra es palindromo**

palabra: palabra ingresada por el usuario.

palabra invertida: misma palabra pero al reves.


**Lenguaje natural**

El programa solicitara una palabra al usuario.

Despues invertira la palabra y comparara si es igual a la original.

Si son iguales mostrara que es un palindromo, si no mostrara que no lo es.

```python
palabra = input("ingrese una palabra:")
palabra_invertida = palabra[::-1]

if palabra == palabra_invertida:
  print("la palabra es un palindromo")
else:
  print("la palabra no es un palindremo")
```
