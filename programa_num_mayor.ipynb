**Definicion de variables**

Num 1:Primer numero ingresado por el usuario.

Num 2:Segundo numero ingresado.

Num 3:Tercer numero ingresado.

Mayor:Variable para almacenar el numero mayor.

**Descripcion en lenguaje natural**

El programa pedira al usuario que escriba tres numeros.

Luego comparara los tres numeros entre si para saber cual es el mayor.

Finalmente mostrara el numero mayor en pantalla.

```python
num1 = float(input("ingrese el primer numero: "))
num2 = float(input("ingrese el segundo numero: "))
num3 = float(input("ingrese el tercer numero: "))

mayor = num1

if num2 > mayor:
  mayor = num2

if num3 > mayor:
  mayor = num3

print("el numero mayor es:", mayor)
```
